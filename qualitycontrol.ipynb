```python
import os
from PIL import Image
import numpy as np

def get_image_sizes(path):
    files = os.listdir(path)
    sizes = []
    for file in files:
        if file.endswith(".jpg"):
            image = Image.open(path + file)
            sizes.append(image.size)
    return sizes

sizes = get_image_sizes("ODIR-5K_Training_Dataset/")
sizes = np.array(sizes)

#give me all possible widhts and heights
print("all possible widths: ", np.unique(sizes[:,0]))
print("all possible heights: ", np.unique(sizes[:,1]))

#which width height combination occur the most
from collections import Counter
Counter(sizes[:,0]).most_common(10)
```
