```python
import pandas as pd
import numpy as np

df = pd.read_csv("ODIR-5K_Training_multipleLR.csv")

search_strings = ['low image quality', 'optic disk photographically invisible', 'lens dust', 'image offset']

for search_string in search_strings:
    df[search_string] = np.where(df['Diagnostic Keywords'].str.contains(search_string), True, False)

df = df[~df[search_strings].any(axis=1)]
df = df.drop(columns=search_strings)

df.to_csv("ODIR-5K_Training_multipleLR_cleaned.csv", index=False)
```
