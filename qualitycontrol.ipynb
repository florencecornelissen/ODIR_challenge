```python
import pandas as pd
import numpy as np

df = pd.read_csv("ODIR-5K_Training_Annotations_LR.csv")
```

```python
df_new = pd.DataFrame(np.repeat(df.values, 2, axis=0))
df_new.columns = df.columns
```

```python
def append_extension(fn, side):
    return f"{fn}_{side}.jpg"

df_new['Side'] = df_new.index % 2
df_new['ID'] = df_new.apply(lambda row: append_extension(row['ID'], 'left' if row['Side'] == 0 else 'right'), axis=1)

df_new.drop(columns=['Side'], inplace=True)

df_new
```
