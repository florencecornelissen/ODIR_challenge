```python
import os
import glob
from PIL import Image
import matplotlib.pyplot as plt

#file_path2 = os.path.abspath(r'C:\Users\elise\OneDrive - Complexit\00_Vrije_Universiteit\AI_for_MI\ODIR_challange\ODIR-5K_Training_Dataset')
file_path2 = os.path.abspath(r'C:\Users\elise\OneDrive - Complexit\00_Vrije_Universiteit\AI_for_MI\Grand_Challenge\ODIR-5K_Training_Dataset')

# Get a list of image file paths 
image_files = glob.glob(f'{file_path2}/*')

# Create subplots for displaying multiple images
num_images = len(image_files)
rows, cols = 2, 5 
fig, axes = plt.subplots(rows, cols, figsize=(12, 6))

# Loop through the first 10 image files and display them
for i, image_path in enumerate(image_files[:10]):
    row, col = i // cols, i % cols
    img = Image.open(image_path)
    axes[row, col].imshow(img)
    axes[row, col].set_title(f'Image {i + 1}')
    axes[row, col].axis('off')

# If there are fewer than 10 images, clear the remaining subplots
for i in range(num_images, rows * cols):
    row, col = i // cols, i % cols
    axes[row, col].axis('off')

#plt.tight_layout()
plt.show()

```

```python
# Create subplots for displaying cropped images
rows, cols = 4, 5  
fig, axes = plt.subplots(rows, cols, figsize=(12, 6))

# Loop through the first 10 image files, crop them, and display
for i, image_path in enumerate(image_files[:20]):
    row, col = i // cols, i % cols
    img = Image.open(image_path)
    img = img.resize((124, 124))  # Resize to (124, 124) --> cropping does not work
    axes[row, col].imshow(img)
    img.show
    
    axes[row, col].set_title(f'Cropped Image {i + 1}')
    axes[row, col].axis('off')

plt.tight_layout()
plt.show()

```
